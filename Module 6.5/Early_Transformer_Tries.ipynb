```python
import os
import re
import string
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# ---------------------------
# Preprocessing Functions
# ---------------------------
def preprocess_text(text):
    """
    Cleans up the text by removing boilerplate, carriage returns,
    fixing quotes, marking original capitals (with '^') and then lowercasing,
    normalizing whitespace, and separating punctuation with spaces.
    """
    text = text.replace("Project Gutenberg", "").replace("Gutenberg", "")
    text = text.replace("\r", "")
    text = text.replace("“", "\"").replace("”", "\"")
    text = re.sub(r"(?<![a-zA-Z])([A-Z])", lambda m: f"^{m.group(0).lower()}", text)
    text = re.sub(r"([A-Z])", lambda m: m.group(0).lower(), text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub("([{}])".format(re.escape(string.punctuation)), r" \1 ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

def postprocess_text(text):
    """
    Restores capitalization by converting caret-marked letters to uppercase.
    """
    return re.sub(r"\^([a-z])", lambda m: m.group(1).upper(), text)

def getMyText():
    """
    Loads the text from a local file or downloads it if necessary.
    Returns the preprocessed text.
    """
    file_name = 'austen.txt'
    file_url = 'https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/austen/austen.txt'
    local_dir = 'saved_files'
    local_path = os.path.join(local_dir, file_name)

    try:
        if not os.path.exists(local_dir):
            os.makedirs(local_dir)
        if os.path.exists(local_path):
            print(f"File '{file_name}' found locally. Using it.")
        else:
            print(f"File '{file_name}' not found locally. Downloading it.")
            downloaded_path = tf.keras.utils.get_file(file_name, file_url)
            with open(downloaded_path, 'rb') as src, open(local_path, 'wb') as dst:
                dst.write(src.read())
        with open(local_path, 'rb') as file:
            text = file.read().decode('utf-8')
        return preprocess_text(text)
    except Exception as e:
        print(f"An error occurred while loading text: {e}")
        return None

# ---------------------------
# Load and Preprocess Text
# ---------------------------
text_data = getMyText()
if text_data is None:
    raise ValueError("Text data could not be loaded.")

print("Length of preprocessed text (characters):", len(text_data))
print("Text snippet (preprocessed):")
print(text_data[:500])
print("----------")

# ---------------------------
# Custom BPE Tokenizer
# ---------------------------
# If needed, install the tokenizers library:
# !pip install tokenizers

from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.normalizers import Sequence, Lowercase, Strip

tokenizer_file = "custom_tokenizer.json"
target_vocab_size = 18286  # ~18K tokens

if os.path.exists(tokenizer_file):
    print("Loading existing BPE tokenizer.")
    tokenizer = Tokenizer.from_file(tokenizer_file)
else:
    print("Training a new BPE tokenizer.")
    tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
    tokenizer.normalizer = Sequence([Lowercase(), Strip()])
    tokenizer.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=target_vocab_size,
                         special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
    tokenizer.train_from_iterator([text_data], trainer=trainer)
    tokenizer.save(tokenizer_file)

vocab = tokenizer.get_vocab()
print("Unique token count in tokenizer vocabulary:", len(vocab))
print("----------")

# Encode full text using the custom BPE tokenizer.
full_encoding = tokenizer.encode(text_data)
full_token_ids = full_encoding.ids
total_tokens = len(full_token_ids)
print(f"Total tokens in full text: {total_tokens}")
print("Sample decoded tokens (first 50 tokens):", tokenizer.decode(full_token_ids[:50]))
print("----------")

# ---------------------------
# Create Sequences for Training
# ---------------------------
sequence_length = 128  # Input sequence length

def create_sequences(token_ids, seq_length):
    sequences = []
    for i in range(0, len(token_ids) - seq_length):
        sequences.append(token_ids[i: i + seq_length + 1])
    return np.array(sequences)

sequences = create_sequences(full_token_ids, sequence_length)
print(f"Total sequences generated: {len(sequences)}")
if sequences.size > 0:
    print("Length of first sequence:", len(sequences[0]))
print("----------")

def split_input_target(seq):
    return seq[:-1], seq[1:]

np.random.shuffle(sequences)
split_idx = int(0.9 * len(sequences))
train_sequences = sequences[:split_idx]
val_sequences = sequences[split_idx:]
print("Training sequences:", len(train_sequences))
print("Validation sequences:", len(val_sequences))

BATCH_SIZE = 64
BUFFER_SIZE = 10000

train_dataset = tf.data.Dataset.from_tensor_slices(train_sequences)
train_dataset = train_dataset.map(lambda seq: split_input_target(seq))
train_dataset = train_dataset.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)\
                                 .cache().prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices(val_sequences)
val_dataset = val_dataset.map(lambda seq: split_input_target(seq))
val_dataset = val_dataset.batch(BATCH_SIZE, drop_remainder=True)\
                         .cache().prefetch(tf.data.AUTOTUNE)

# ---------------------------
# Helper Functions for Rotary Positional Embeddings (RoPE)
# ---------------------------
def get_rotary_embeddings(seq_len, head_dim):
    """
    Computes sin and cos embeddings for RoPE.
    Assumes head_dim is even.
    """
    position = tf.range(seq_len, dtype=tf.float32)  # (seq_len,)
    inv_freq = 1.0 / (10000 ** (tf.range(0, head_dim, 2, dtype=tf.float32) / tf.cast(head_dim, tf.float32)))
    sinusoid_inp = tf.einsum('i,j->ij', position, inv_freq)  # (seq_len, head_dim//2)
    sin = tf.sin(sinusoid_inp)  # (seq_len, head_dim//2)
    cos = tf.cos(sinusoid_inp)  # (seq_len, head_dim//2)
    return sin, cos

def apply_rope(x, sin, cos):
    """
    Applies rotary positional embeddings to tensor x.
    x: (batch, seq_len, num_heads, head_dim)
    sin, cos: (seq_len, head_dim//2)
    """
    head_dim = tf.shape(x)[-1]
    # Reshape last dimension to (head_dim//2, 2)
    x = tf.reshape(x, tf.concat([tf.shape(x)[:-1], [head_dim // 2, 2]], axis=0))
    x1, x2 = x[..., 0], x[..., 1]
    # Expand sin and cos to broadcast: (1, seq_len, 1, head_dim//2)
    sin = tf.expand_dims(tf.expand_dims(sin, axis=0), axis=2)
    cos = tf.expand_dims(tf.expand_dims(cos, axis=0), axis=2)
    # Apply rotation.
    x_rotated_first = x1 * cos - x2 * sin
    x_rotated_second = x1 * sin + x2 * cos
    x = tf.stack([x_rotated_first, x_rotated_second], axis=-1)
    # Reshape back to (batch, seq_len, num_heads, head_dim)
    return tf.reshape(x, tf.concat([tf.shape(x)[:-2], [head_dim]], axis=0))

# ---------------------------
# Custom Layers: RMSNorm and TiedDense
# ---------------------------
class RMSNorm(tf.keras.layers.Layer):
    """Root Mean Square Normalization."""
    def __init__(self, epsilon=1e-8, **kwargs):
        super(RMSNorm, self).__init__(**kwargs)
        self.epsilon = epsilon

    def build(self, input_shape):
        self.gamma = self.add_weight(
            name="gamma", shape=input_shape[-1:], initializer="ones", trainable=True
        )
        super(RMSNorm, self).build(input_shape)

    def call(self, inputs):
        rms = tf.sqrt(tf.reduce_mean(tf.square(inputs), axis=-1, keepdims=True) + self.epsilon)
        return inputs * self.gamma / rms

class TiedDense(tf.keras.layers.Layer):
    """
    A Dense layer that ties its weights to the provided embedding layer.
    It multiplies the input by the transpose of the embedding matrix.
    """
    def __init__(self, tied_to, **kwargs):
        super(TiedDense, self).__init__(**kwargs)
        self.tied_to = tied_to

    def call(self, inputs):
        return tf.matmul(inputs, self.tied_to.embeddings, transpose_b=True)

# ---------------------------
# Custom Rotary Self-Attention Layer
# ---------------------------
class RotarySelfAttention(tf.keras.layers.Layer):
    """
    Multi-head self-attention layer with Rotary Positional Embeddings (RoPE).
    Computes query, key, and value projections, applies RoPE to query and key,
    and then computes scaled dot-product attention.
    """
    def __init__(self, embed_dim, num_heads, dropout_rate=0.1, **kwargs):
        super(RotarySelfAttention, self).__init__(**kwargs)
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads
        if self.head_dim * num_heads != embed_dim:
            raise ValueError("embed_dim must be divisible by num_heads")
        self.dropout_rate = dropout_rate

    def build(self, input_shape):
        self.query_dense = tf.keras.layers.Dense(self.embed_dim)
        self.key_dense = tf.keras.layers.Dense(self.embed_dim)
        self.value_dense = tf.keras.layers.Dense(self.embed_dim)
        self.out_dense = tf.keras.layers.Dense(self.embed_dim)
        self.dropout = tf.keras.layers.Dropout(self.dropout_rate)
        super(RotarySelfAttention, self).build(input_shape)

    def call(self, inputs, training=False, use_causal_mask=True):
        # inputs: (batch, seq_len, embed_dim)
        batch_size = tf.shape(inputs)[0]
        seq_len = tf.shape(inputs)[1]

        # Linear projections.
        query = self.query_dense(inputs)  # (batch, seq_len, embed_dim)
        key = self.key_dense(inputs)
        value = self.value_dense(inputs)

        # Reshape to (batch, seq_len, num_heads, head_dim)
        query = tf.reshape(query, (batch_size, seq_len, self.num_heads, self.head_dim))
        key = tf.reshape(key, (batch_size, seq_len, self.num_heads, self.head_dim))
        value = tf.reshape(value, (batch_size, seq_len, self.num_heads, self.head_dim))

        # Compute RoPE sin and cos embeddings.
        sin, cos = get_rotary_embeddings(seq_len, self.head_dim)
        # Apply RoPE to query and key.
        query = apply_rope(query, sin, cos)
        key = apply_rope(key, sin, cos)

        # Transpose to (batch, num_heads, seq_len, head_dim)
        query = tf.transpose(query, perm=[0, 2, 1, 3])
        key = tf.transpose(key, perm=[0, 2, 1, 3])
        value = tf.transpose(value, perm=[0, 2, 1, 3])

        # Scale query.
        scaling = tf.cast(self.head_dim, tf.float32) ** -0.5
        query = query * scaling

        # Compute attention scores.
        attn_logits = tf.matmul(query, key, transpose_b=True)  # (batch, num_heads, seq_len, seq_len)
        if use_causal_mask:
            # Create a causal mask.
            mask = tf.linalg.band_part(tf.ones((seq_len, seq_len)), -1, 0)
            mask = tf.reshape(mask, (1, 1, seq_len, seq_len))
            attn_logits = attn_logits * mask + -1e9 * (1 - mask)

        # Apply softmax to get attention weights.
        attn_weights = tf.nn.softmax(attn_logits, axis=-1)
        attn_weights = self.dropout(attn_weights, training=training)

        # Compute attention output.
        attn_output = tf.matmul(attn_weights, value)  # (batch, num_heads, seq_len, head_dim)
        attn_output = tf.transpose(attn_output, perm=[0, 2, 1, 3])
        attn_output = tf.reshape(attn_output, (batch_size, seq_len, self.embed_dim))
        output = self.out_dense(attn_output)
        return output

# ---------------------------
# Transformer Block Using RoPE
# ---------------------------
class TransformerBlock(tf.keras.layers.Layer):
    """
    Transformer block with Rotary Self-Attention, feed-forward network,
    dropout, and RMSNorm normalization.
    """
    def __init__(self, embed_dim, num_heads, ff_dim, dropout_rate=0.1, **kwargs):
        super(TransformerBlock, self).__init__(**kwargs)
        self.attention = RotarySelfAttention(embed_dim, num_heads, dropout_rate)
        self.dropout1 = tf.keras.layers.Dropout(dropout_rate)
        self.norm1 = RMSNorm()
        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(ff_dim, activation=tf.nn.gelu),
            tf.keras.layers.Dense(embed_dim)
        ])
        self.dropout2 = tf.keras.layers.Dropout(dropout_rate)
        self.norm2 = RMSNorm()

    def call(self, inputs, training=False):
        attn_output = self.attention(inputs, training=training, use_causal_mask=True)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.norm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.norm2(out1 + ffn_output)

# ---------------------------
# Build the Transformer Model with Weight Tying
# ---------------------------
def create_transformer_model(vocab_size, sequence_length, embed_dim, num_heads, ff_dim, num_layers, dropout_rate=0.1):
    inputs = tf.keras.Input(shape=(sequence_length,), dtype=tf.int32)

    # Shared token embedding layer for weight tying.
    token_embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, name="token_embedding")
    x = token_embedding(inputs)  # (batch, seq_len, embed_dim)
    # Note: We omit a separate positional embedding layer since RoPE is applied in attention.

    # Pass through a stack of transformer blocks.
    for i in range(num_layers):
        x = TransformerBlock(embed_dim, num_heads, ff_dim, dropout_rate, name=f"transformer_block_{i}")(x)

    # Final normalization for training stability.
    x = tf.keras.layers.LayerNormalization(name="final_layer_norm")(x)

    # Output projection with weight tying.
    logits = TiedDense(token_embedding, name="output_projection")(x)

    model = tf.keras.Model(inputs=inputs, outputs=logits)
    return model

vocab_size_model = len(tokenizer.get_vocab())  # ~18286 tokens.
embed_dim = 352     # New embedding dimension.
num_heads = 8       # 8 attention heads.
ff_dim = embed_dim * 4  # Feed-forward expansion (1:4 ratio), i.e. 1408.
num_layers = 8      # Increased transformer depth.
dropout_rate = 0.1

transformer_model = create_transformer_model(vocab_size_model, sequence_length, embed_dim, num_heads, ff_dim, num_layers, dropout_rate)

# ---------------------------
# Learning Rate Schedule: Warm-up Cosine Decay
# ---------------------------
class WarmUpCosineDecay(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, initial_lr, total_steps, warmup_steps, alpha=0.0):
        super(WarmUpCosineDecay, self).__init__()
        self.initial_lr = initial_lr
        self.total_steps = total_steps
        self.warmup_steps = warmup_steps
        self.alpha = alpha

    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        warmup_lr = self.initial_lr * step / tf.cast(self.warmup_steps, tf.float32)
        cosine_steps = tf.maximum(step - tf.cast(self.warmup_steps, tf.float32), 0.0)
        total_cosine_steps = tf.maximum(tf.cast(self.total_steps - self.warmup_steps, tf.float32), 1.0)
        cosine_decay = 0.5 * (1 + tf.cos(tf.constant(np.pi, dtype=tf.float32) * cosine_steps / total_cosine_steps))
        decayed_lr = self.alpha * self.initial_lr + (1 - self.alpha) * self.initial_lr * cosine_decay
        return tf.where(step < tf.cast(self.warmup_steps, tf.float32), warmup_lr, decayed_lr)

num_epochs = 7
steps_per_epoch = len(train_dataset)
total_steps = num_epochs * steps_per_epoch
warmup_steps = int(0.1 * total_steps)
initial_lr = 1e-3

lr_schedule = WarmUpCosineDecay(initial_lr, total_steps, warmup_steps, alpha=0.0)
optimizer = tf.keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=1e-4)

transformer_model.compile(optimizer=optimizer,
                          loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                          metrics=['accuracy'])
transformer_model.summary()

# ---------------------------
# Checkpointing and Resuming Training
# ---------------------------
checkpoint_dir = 'checkpoints'
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)
checkpoint_path = os.path.join(checkpoint_dir, "cp-{epoch:04d}.weights.h5")

cp_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    save_weights_only=True,
    verbose=1,
    save_freq='epoch'
)

latest_checkpoint = tf.train.latest_checkpoint(checkpoint_dir)
initial_epoch = 0
if latest_checkpoint:
    print("Loading weights from", latest_checkpoint)
    transformer_model.load_weights(latest_checkpoint)
    initial_epoch = int(latest_checkpoint.split('-')[1].split('.')[0])
    print("Resuming training from epoch", initial_epoch)

# ---------------------------
# Train the Transformer Model
# ---------------------------
try:
    history = transformer_model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=num_epochs,
        initial_epoch=initial_epoch,
        callbacks=[cp_callback]
    )
except Exception as e:
    print("An error occurred during training:", e)

# ---------------------------
# Plot Training Metrics
# ---------------------------
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], marker='o', label='Training Loss')
plt.plot(history.history['val_loss'], marker='o', label='Validation Loss')
plt.title('Training vs. Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], marker='o', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], marker='o', label='Validation Accuracy')
plt.title('Training vs. Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

train_perplexity = np.exp(history.history['loss'])
val_perplexity = np.exp(history.history['val_loss'])
plt.figure(figsize=(10, 5))
plt.plot(train_perplexity, marker='o', label='Training Perplexity')
plt.plot(val_perplexity, marker='o', label='Validation Perplexity')
plt.title('Training vs. Validation Perplexity')
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# Text Generation Utilities with Softmax during Inference
# ---------------------------
def generate_text(model, prompt, num_generate=100):
    """
    Given a prompt, generate additional tokens using the trained model.
    The prompt is preprocessed and encoded using the custom BPE tokenizer.
    Softmax is applied to the logits during inference.
    """
    processed_prompt = preprocess_text(prompt)
    encoding = tokenizer.encode(processed_prompt)
    input_ids = encoding.ids
    if len(input_ids) < sequence_length:
        input_ids = [0] * (sequence_length - len(input_ids)) + input_ids
    else:
        input_ids = input_ids[-sequence_length:]
    input_ids = np.expand_dims(input_ids, 0)

    generated = []
    for _ in range(num_generate):
        predictions = model.predict(input_ids)
        logits = predictions[0, -1, :]  # logits for the last token position
        # Explicitly apply softmax to get probability distribution.
        probs = tf.nn.softmax(logits).numpy()
        # Choose the token with highest probability.
        predicted_id = int(tf.argmax(probs).numpy())
        generated.append(predicted_id)
        input_ids = np.concatenate([input_ids, np.array([[predicted_id]])], axis=1)
        input_ids = input_ids[:, -sequence_length:]
    return postprocess_text(tokenizer.decode(generated))

# ---------------------------
# Generate Sample Text
# ---------------------------
random_prompt = "The world seemed like such a peaceful place until the magic tree was discovered in London."
generated_sample = generate_text(transformer_model, random_prompt, num_generate=100)
print("\nGenerated Text Sample:")
print(random_prompt ," ", generated_sample)
```

```python
import os
import re
import string
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.normalizers import Sequence, Lowercase, Strip

# ---------------------------
# Preprocessing Functions
# ---------------------------
def preprocess_text(text):
    text = text.replace("Project Gutenberg", "").replace("Gutenberg", "")
    text = text.replace("\r", "")
    text = text.replace("“", "\"").replace("”", "\"")
    # Mark capital letters with caret and then lowercase
    text = re.sub(r"(?<![a-zA-Z])([A-Z])", lambda m: f"^{m.group(0).lower()}", text)
    text = re.sub(r"([A-Z])", lambda m: m.group(0).lower(), text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub("([{}])".format(re.escape(string.punctuation)), r" \1 ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

def postprocess_text(text):
    # Restore capitalization and fix punctuation spacing.
    text = re.sub(r'\^([a-z])', lambda m: m.group(1).upper(), text)
    text = re.sub(r'\s+([,.!?;:])', r'\1', text)
    return text.strip()

def get_text():
    file_name = 'austen.txt'
    file_url = 'https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/austen/austen.txt'
    local_dir = 'saved_files'
    local_path = os.path.join(local_dir, file_name)
    try:
        if not os.path.exists(local_dir):
            os.makedirs(local_dir)
        if not os.path.exists(local_path):
            print(f"Downloading {file_name}...")
            downloaded_path = tf.keras.utils.get_file(file_name, file_url)
            with open(downloaded_path, 'rb') as src, open(local_path, 'wb') as dst:
                dst.write(src.read())
        with open(local_path, 'rb') as file:
            text = file.read().decode('utf-8')
        return preprocess_text(text)
    except Exception as e:
        print("Error loading text:", e)
        return None

text_data = get_text()
if text_data is None:
    raise ValueError("Text data could not be loaded.")

print("Length of preprocessed text (characters):", len(text_data))
print(text_data[:300])
print("----------")

# ---------------------------
# Train a New BPE Tokenizer (Self-Contained)
# ---------------------------
print("Training a new BPE tokenizer.")
target_vocab_size = 18286  # ~18K tokens
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
tokenizer.normalizer = Sequence([Lowercase(), Strip()])
tokenizer.pre_tokenizer = Whitespace()
trainer = BpeTrainer(vocab_size=target_vocab_size, special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
tokenizer.train_from_iterator([text_data], trainer=trainer)
vocab = tokenizer.get_vocab()
print("Tokenizer vocab size:", len(vocab))
print("----------")

# ---------------------------
# Tokenize Full Text and Create Sequences
# ---------------------------
full_encoding = tokenizer.encode(text_data)
full_token_ids = full_encoding.ids
print("Total tokens in full text:", len(full_token_ids))
print("Sample tokens:", tokenizer.decode(full_token_ids[:50]))
print("----------")

sequence_length = 128
def create_sequences(token_ids, seq_length):
    sequences = []
    for i in range(0, len(token_ids) - seq_length):
        sequences.append(token_ids[i: i + seq_length + 1])
    return np.array(sequences)

sequences = create_sequences(full_token_ids, sequence_length)
print("Total sequences:", len(sequences))
np.random.shuffle(sequences)
split_idx = int(0.9 * len(sequences))
train_sequences = sequences[:split_idx]
val_sequences = sequences[split_idx:]

def split_input_target(seq):
    return seq[:-1], seq[1:]

BATCH_SIZE = 64
BUFFER_SIZE = 10000

train_dataset = tf.data.Dataset.from_tensor_slices(train_sequences)
train_dataset = train_dataset.map(lambda seq: split_input_target(seq))\
                             .shuffle(BUFFER_SIZE)\
                             .batch(BATCH_SIZE, drop_remainder=True)\
                             .cache()\
                             .prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices(val_sequences)
val_dataset = val_dataset.map(lambda seq: split_input_target(seq))\
                           .batch(BATCH_SIZE, drop_remainder=True)\
                           .cache()\
                           .prefetch(tf.data.AUTOTUNE)

# ---------------------------
# RoPE Helper Functions
# ---------------------------
def get_rotary_embeddings(seq_len, head_dim):
    position = tf.range(seq_len, dtype=tf.float32)
    inv_freq = 1.0 / (10000 ** (tf.range(0, head_dim, 2, dtype=tf.float32) / tf.cast(head_dim, tf.float32)))
    sinusoid_inp = tf.einsum('i,j->ij', position, inv_freq)
    sin = tf.sin(sinusoid_inp)
    cos = tf.cos(sinusoid_inp)
    return sin, cos

def apply_rope(x, sin, cos):
    head_dim = tf.shape(x)[-1]
    x = tf.reshape(x, tf.concat([tf.shape(x)[:-1], [head_dim // 2, 2]], axis=0))
    x1, x2 = x[..., 0], x[..., 1]
    sin = tf.expand_dims(tf.expand_dims(sin, axis=0), axis=2)
    cos = tf.expand_dims(tf.expand_dims(cos, axis=0), axis=2)
    x_rotated_first = x1 * cos - x2 * sin
    x_rotated_second = x1 * sin + x2 * cos
    x = tf.stack([x_rotated_first, x_rotated_second], axis=-1)
    return tf.reshape(x, tf.concat([tf.shape(x)[:-2], [head_dim]], axis=0))

# ---------------------------
# Custom Layers: RMSNorm, TiedDense, and Rotary Self-Attention
# ---------------------------
class RMSNorm(tf.keras.layers.Layer):
    def __init__(self, epsilon=1e-8, **kwargs):
        super(RMSNorm, self).__init__(**kwargs)
        self.epsilon = epsilon
    def build(self, input_shape):
        self.gamma = self.add_weight(name="gamma", shape=input_shape[-1:], initializer="ones", trainable=True)
        super(RMSNorm, self).build(input_shape)
    def call(self, inputs):
        rms = tf.sqrt(tf.reduce_mean(tf.square(inputs), axis=-1, keepdims=True) + self.epsilon)
        return inputs * self.gamma / rms

class TiedDense(tf.keras.layers.Layer):
    def __init__(self, tied_to, **kwargs):
        super(TiedDense, self).__init__(**kwargs)
        self.tied_to = tied_to
    def call(self, inputs):
        return tf.matmul(inputs, self.tied_to.embeddings, transpose_b=True)

class RotarySelfAttention(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, dropout_rate=0.1, **kwargs):
        super(RotarySelfAttention, self).__init__(**kwargs)
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads
        if self.head_dim * num_heads != embed_dim:
            raise ValueError("embed_dim must be divisible by num_heads")
        self.dropout_rate = dropout_rate
    def build(self, input_shape):
        self.query_dense = tf.keras.layers.Dense(self.embed_dim)
        self.key_dense = tf.keras.layers.Dense(self.embed_dim)
        self.value_dense = tf.keras.layers.Dense(self.embed_dim)
        self.out_dense = tf.keras.layers.Dense(self.embed_dim)
        self.dropout = tf.keras.layers.Dropout(self.dropout_rate)
        super(RotarySelfAttention, self).build(input_shape)
    def call(self, inputs, training=False, use_causal_mask=True):
        batch_size = tf.shape(inputs)[0]
        seq_len = tf.shape(inputs)[1]
        query = self.query_dense(inputs)
        key = self.key_dense(inputs)
        value = self.value_dense(inputs)
        query = tf.reshape(query, (batch_size, seq_len, self.num_heads, self.head_dim))
        key = tf.reshape(key, (batch_size, seq_len, self.num_heads, self.head_dim))
        value = tf.reshape(value, (batch_size, seq_len, self.num_heads, self.head_dim))
        sin, cos = get_rotary_embeddings(seq_len, self.head_dim)
        query = apply_rope(query, sin, cos)
        key = apply_rope(key, sin, cos)
        query = tf.transpose(query, perm=[0, 2, 1, 3])
        key = tf.transpose(key, perm=[0, 2, 1, 3])
        value = tf.transpose(value, perm=[0, 2, 1, 3])
        scaling = tf.cast(self.head_dim, tf.float32) ** -0.5
        query = query * scaling
        attn_logits = tf.matmul(query, key, transpose_b=True)
        if use_causal_mask:
            mask = tf.linalg.band_part(tf.ones((seq_len, seq_len)), -1, 0)
            mask = tf.reshape(mask, (1, 1, seq_len, seq_len))
            attn_logits = attn_logits * mask + -1e9 * (1 - mask)
        attn_weights = tf.nn.softmax(attn_logits, axis=-1)
        attn_weights = self.dropout(attn_weights, training=training)
        attn_output = tf.matmul(attn_weights, value)
        attn_output = tf.transpose(attn_output, perm=[0, 2, 1, 3])
        attn_output = tf.reshape(attn_output, (batch_size, seq_len, self.embed_dim))
        return self.out_dense(attn_output)

class TransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout_rate=0.1, **kwargs):
        super(TransformerBlock, self).__init__(**kwargs)
        self.attention = RotarySelfAttention(embed_dim, num_heads, dropout_rate)
        self.dropout1 = tf.keras.layers.Dropout(dropout_rate)
        self.norm1 = RMSNorm()
        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(ff_dim, activation=tf.nn.gelu),
            tf.keras.layers.Dense(embed_dim)
        ])
        self.dropout2 = tf.keras.layers.Dropout(dropout_rate)
        self.norm2 = RMSNorm()
    def call(self, inputs, training=False):
        attn_output = self.attention(inputs, training=training, use_causal_mask=True)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.norm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.norm2(out1 + ffn_output)

def create_transformer_model(vocab_size, sequence_length, embed_dim, num_heads, ff_dim, num_layers, dropout_rate=0.1):
    inputs = tf.keras.Input(shape=(sequence_length,), dtype=tf.int32)
    token_embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, name="token_embedding")
    x = token_embedding(inputs)
    for i in range(num_layers):
        x = TransformerBlock(embed_dim, num_heads, ff_dim, dropout_rate, name=f"transformer_block_{i}")(x)
    x = tf.keras.layers.LayerNormalization(name="final_layer_norm")(x)
    logits = TiedDense(token_embedding, name="output_projection")(x)
    return tf.keras.Model(inputs=inputs, outputs=logits)

vocab_size_model = len(vocab)  # ~18K tokens
embed_dim = 400   # Changed to 400 dimensions
num_heads = 8     # 400/8 = 50 per head
ff_dim = embed_dim * 4  # 1600
num_layers = 8
dropout_rate = 0.1

transformer_model = create_transformer_model(vocab_size_model, sequence_length, embed_dim, num_heads, ff_dim, num_layers, dropout_rate)
transformer_model.summary()

# ---------------------------
# Learning Rate Schedule
# ---------------------------
class WarmUpCosineDecay(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, initial_lr, total_steps, warmup_steps, alpha=0.0):
        super(WarmUpCosineDecay, self).__init__()
        self.initial_lr = initial_lr
        self.total_steps = total_steps
        self.warmup_steps = warmup_steps
        self.alpha = alpha
    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        warmup_lr = self.initial_lr * step / tf.cast(self.warmup_steps, tf.float32)
        cosine_steps = tf.maximum(step - tf.cast(self.warmup_steps, tf.float32), 0.0)
        total_cosine_steps = tf.maximum(tf.cast(self.total_steps - self.warmup_steps, tf.float32), 1.0)
        cosine_decay = 0.5 * (1 + tf.cos(np.pi * cosine_steps / total_cosine_steps))
        decayed_lr = self.alpha * self.initial_lr + (1 - self.alpha) * self.initial_lr * cosine_decay
        return tf.where(step < tf.cast(self.warmup_steps, tf.float32), warmup_lr, decayed_lr)

num_epochs = 7
steps_per_epoch = len(train_dataset)
total_steps = num_epochs * steps_per_epoch
warmup_steps = int(0.1 * total_steps)
initial_lr = 1e-3

lr_schedule = WarmUpCosineDecay(initial_lr, total_steps, warmup_steps, alpha=0.0)
optimizer = tf.keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=1e-4)

transformer_model.compile(optimizer=optimizer,
                          loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                          metrics=['accuracy'])

# ---------------------------
# Checkpointing Setup with tf.train.Checkpoint & Manager
# ---------------------------
checkpoint_dir = 'checkpoints'
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)
global_epoch = tf.Variable(0, trainable=False, dtype=tf.int64)
checkpoint = tf.train.Checkpoint(model=transformer_model, optimizer=optimizer, epoch=global_epoch)
manager = tf.train.CheckpointManager(checkpoint, checkpoint_dir, max_to_keep=5)

initial_epoch = 0
if manager.latest_checkpoint:
    checkpoint.restore(manager.latest_checkpoint)
    print("Restored from", manager.latest_checkpoint)
    initial_epoch = int(global_epoch.numpy())

class EpochUpdater(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        global_epoch.assign(epoch + 1)
        manager.save()

# ---------------------------
# Train the Transformer Model
# ---------------------------
history = transformer_model.fit(train_dataset,
                                  validation_data=val_dataset,
                                  epochs=num_epochs,
                                  initial_epoch=initial_epoch,
                                  callbacks=[EpochUpdater()])

# ---------------------------
# Plot Training Metrics
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], marker='o', label='Training Loss')
plt.plot(history.history['val_loss'], marker='o', label='Validation Loss')
plt.title('Training vs. Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(history.history['accuracy'], marker='o', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], marker='o', label='Validation Accuracy')
plt.title('Training vs. Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# ---------------------------
# Text Generation with Softmax Sampling (Temperature = 1)
# ---------------------------
def generate_text(model, prompt, num_generate=100, temperature=1.0):
    processed_prompt = preprocess_text(prompt)
    encoding = tokenizer.encode(processed_prompt)
    input_ids = encoding.ids
    if len(input_ids) < sequence_length:
        input_ids = [0]*(sequence_length - len(input_ids)) + input_ids
    else:
        input_ids = input_ids[-sequence_length:]
    input_ids = np.expand_dims(input_ids, 0)
    generated = []
    for _ in range(num_generate):
        predictions = model.predict(input_ids)
        # Divide logits by temperature (here temperature is fixed to 1)
        logits = predictions[0, -1, :] / temperature
        probs = tf.nn.softmax(logits).numpy()
        predicted_id = np.random.choice(len(probs), p=probs)
        generated.append(predicted_id)
        input_ids = np.concatenate([input_ids, np.array([[predicted_id]])], axis=1)
        input_ids = input_ids[:, -sequence_length:]
    return postprocess_text(tokenizer.decode(generated))

# ---------------------------
# Generate Sample Text
# ---------------------------
sample_prompt = "The world seemed like such a peaceful place until the magic tree was discovered in London."
generated_sample = generate_text(transformer_model, sample_prompt, num_generate=100, temperature=1.0)
print("\nGenerated Text Sample:")
print(sample_prompt, "\n", generated_sample)
```
